```python
np.random.seed(42)
random.seed(42)

df = pd.read_csv('Airplane.csv')

# Remove unnecessary columns like 'Unnamed: 0' and 'id'
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])
if 'id' in df.columns:
    df = df.drop(columns=['id'])

print(f"Dataset shape after removing unnecessary columns: {df.shape}")
print(f"Number of samples: {df.shape[0]}")
print(f"Number of features: {df.shape[1]}")


# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

## Pre-processing

```python
# Process data - encode categorical variables and handle missing values
# Create a copy of the dataframe to process
full_data = df.copy()

# Handle the target variable
full_data['satisfaction'] = full_data['satisfaction'].map({"neutral or dissatisfied": 0, "satisfied": 1}).astype(float) 

# Gender
full_data['Gender'] = full_data['Gender'].fillna("unknown") 
full_data['Gender'] = full_data['Gender'].map({"Female": 0, "Male": 1}).astype(float)

# Customer Type
full_data['Customer Type'] = full_data['Customer Type'].fillna("unknown")
full_data['Customer Type'] = full_data['Customer Type'].map({"Loyal Customer": 0, "disloyal Customer": 1}).astype(float)

# Age
full_data['Age'] = full_data['Age'].fillna(full_data['Age'].median())
full_data.loc[full_data['Age'] <= 16 , 'Age'] = 0 
full_data.loc[(full_data['Age'] > 16) & (full_data['Age'] <= 32), 'Age'] = 1
full_data.loc[(full_data['Age'] > 32) & (full_data['Age'] <= 48) , 'Age'] = 2
full_data.loc[(full_data['Age'] > 48) & (full_data['Age'] <= 64) , 'Age'] = 3
full_data.loc[(full_data['Age'] > 64) & (full_data['Age'] <= 85) , 'Age'] = 4

# Type of Travel
full_data['Type of Travel'] = full_data['Type of Travel'].fillna("unknown")
full_data['Type of Travel'] = full_data['Type of Travel'].map({"Personal Travel": 0, "Business travel": 1}).astype(float)

# Class
full_data['Class'] = full_data['Class'].fillna("unknown")
full_data['Class'] = full_data['Class'].map({"Business": 0, "Eco": 1, "Eco Plus": 2}).astype(float)

# Flight Distance
full_data['Flight Distance'] = full_data['Flight Distance'].fillna(full_data['Flight Distance'].median())
full_data.loc[full_data['Flight Distance'] <= 1000 , 'Flight Distance'] = 0 
full_data.loc[(full_data['Flight Distance'] > 1000) & (full_data['Flight Distance'] <= 2000), 'Flight Distance'] = 1
full_data.loc[(full_data['Flight Distance'] > 2000) & (full_data['Flight Distance'] <= 3000) , 'Flight Distance'] = 2
full_data.loc[(full_data['Flight Distance'] > 3000) & (full_data['Flight Distance'] <= 4000) , 'Flight Distance'] = 3
full_data.loc[(full_data['Flight Distance'] > 4000) & (full_data['Flight Distance'] <= 5000) , 'Flight Distance'] = 4

# Inflight wifi service
full_data['Inflight wifi service'] = full_data['Inflight wifi service'].fillna(full_data['Inflight wifi service'].median())

# Departure/Arrival time convenient
full_data['Departure/Arrival time convenient'] = full_data['Departure/Arrival time convenient'].fillna(full_data['Departure/Arrival time convenient'].median())

# Ease of Online booking
full_data['Ease of Online booking'] = full_data['Ease of Online booking'].fillna(full_data['Ease of Online booking'].median())

# Gate location
full_data['Gate location'] = full_data['Gate location'].fillna(full_data['Gate location'].median())

# Food and drink
full_data['Food and drink'] = full_data['Food and drink'].fillna(full_data['Food and drink'].median())

# Online boarding
full_data['Online boarding'] = full_data['Online boarding'].fillna(full_data['Online boarding'].median())

# Seat comfort
full_data['Seat comfort'] = full_data['Seat comfort'].fillna(full_data['Seat comfort'].median())

# Inflight entertainment
full_data['Inflight entertainment'] = full_data['Inflight entertainment'].fillna(full_data['Inflight entertainment'].median())

# On-board service
full_data['On-board service'] = full_data['On-board service'].fillna(full_data['On-board service'].median())

# Leg room service
full_data['Leg room service'] = full_data['Leg room service'].fillna(full_data['Leg room service'].median())

# Baggage handling
full_data['Baggage handling'] = full_data['Baggage handling'].fillna(full_data['Baggage handling'].median())

# Checkin service
full_data['Checkin service'] = full_data['Checkin service'].fillna(full_data['Checkin service'].median())

# Inflight service
full_data['Inflight service'] = full_data['Inflight service'].fillna(full_data['Inflight service'].median())

# Cleanliness
full_data['Cleanliness'] = full_data['Cleanliness'].fillna(full_data['Cleanliness'].median())

# Departure Delay in Minutes
full_data['Departure Delay in Minutes'] = full_data['Departure Delay in Minutes'].fillna(full_data['Departure Delay in Minutes'].median())
full_data.loc[full_data['Departure Delay in Minutes'] <= 300 , 'Departure Delay in Minutes'] = 0 
full_data.loc[(full_data['Departure Delay in Minutes'] > 300) & (full_data['Departure Delay in Minutes'] <= 600), 'Departure Delay in Minutes'] = 1
full_data.loc[(full_data['Departure Delay in Minutes'] > 600) & (full_data['Departure Delay in Minutes'] <= 900), 'Departure Delay in Minutes'] = 2
full_data.loc[(full_data['Departure Delay in Minutes'] > 900) & (full_data['Departure Delay in Minutes'] <= 1200), 'Departure Delay in Minutes'] = 3
full_data.loc[(full_data['Departure Delay in Minutes'] > 1200) & (full_data['Departure Delay in Minutes'] <= 1592), 'Departure Delay in Minutes'] = 4

# Arrival Delay in Minutes
full_data['Arrival Delay in Minutes'] = full_data['Arrival Delay in Minutes'].fillna(full_data['Arrival Delay in Minutes'].median())
full_data.loc[full_data['Arrival Delay in Minutes'] <= 300 , 'Arrival Delay in Minutes'] = 0 
full_data.loc[(full_data['Arrival Delay in Minutes'] > 300) & (full_data['Arrival Delay in Minutes'] <= 600), 'Arrival Delay in Minutes'] = 1
full_data.loc[(full_data['Arrival Delay in Minutes'] > 600) & (full_data['Arrival Delay in Minutes'] <= 900), 'Arrival Delay in Minutes'] = 2
full_data.loc[(full_data['Arrival Delay in Minutes'] > 900) & (full_data['Arrival Delay in Minutes'] <= 1200), 'Arrival Delay in Minutes'] = 3
full_data.loc[(full_data['Arrival Delay in Minutes'] > 1200) & (full_data['Arrival Delay in Minutes'] <= 1584), 'Arrival Delay in Minutes'] = 4
```

```python
# Check for any remaining null values
print("\nRemaining missing values after processing:")
print(full_data.isnull().sum())


X = full_data.drop(columns=['satisfaction'])  # Features
y = full_data['satisfaction']  # Target

# Split into training and test sets with at least 6000 training samples and 2000 test samples
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=2000, random_state=42)

# Ensure we have at least 6000 training samples
if len(X_train) > 6000:
    X_train = X_train.iloc[:6000]
    y_train = y_train.iloc[:6000]

print(f"\nTraining samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
```

## Decision Tree functions

```python
# Class definitions for the custom decision tree
class Node:
    def __init__(self, feature_index=None, thresholds=None, children=None, info_gain=None, value=None):
        self.feature_index = feature_index
        self.thresholds = thresholds
        self.children = children
        self.info_gain = info_gain
        self.value = value

class DecisionTreeClassifier:
    def __init__(self, min_sample_split=2, max_depth=2, num_splits=3, criterion="entropy"):
        self.root = None
        self.min_sample_split = min_sample_split
        self.max_depth = max_depth
        self.num_splits = num_splits
        self.criterion = criterion
        self.feature_names = None

    def build_tree(self, dataset, curr_depth=0):
        X, Y = dataset[:, :-1], dataset[:, -1]
        num_samples, num_features = np.shape(X)

        if num_samples >= self.min_sample_split and curr_depth <= self.max_depth:
            best_split = self.get_best_split(dataset, num_samples, num_features)
            if best_split["info_gain"] > 0:
                children = [self.build_tree(subset, curr_depth + 1) for subset in best_split["datasets"]]
                return Node(feature_index=best_split["feature_index"], thresholds=best_split["thresholds"], 
                            children=children, info_gain=best_split["info_gain"])

        leaf_value = self.calculate_leaf_value(Y)
        return Node(value=leaf_value)

    def get_best_split(self, dataset, num_samples, num_features):
        best_split = {"info_gain": -float("inf")}
        
        for feature_index in range(num_features):
            feature_values = dataset[:, feature_index]
            possible_thresholds = np.unique(feature_values)
            
            # If there are too many unique values, take a sample
            if len(possible_thresholds) > self.num_splits:
                possible_thresholds = np.quantile(possible_thresholds, 
                                               np.linspace(0, 1, self.num_splits + 1)[1:-1])
            
            # If we don't have enough unique values, skip this feature
            if len(possible_thresholds) < 1:
                continue
                
            # Try different combinations of thresholds
            for i in range(1, min(len(possible_thresholds), self.num_splits) + 1):
                for thresholds in [sorted(thresh) for thresh in 
                                  [list(x) for x in np.array(np.meshgrid(*[possible_thresholds] * i)).T.reshape(-1, i)]]:
                    if len(set(thresholds)) != len(thresholds):  # Skip duplicates
                        continue
                    
                    # Split the dataset based on the thresholds
                    datasets = self.split(dataset, feature_index, thresholds)
                    
                    # Skip if any child has no samples
                    if any(len(subset) == 0 for subset in datasets):
                        continue
                    
                    # Calculate information gain
                    info_gain = self.information_gain(dataset[:, -1], [subset[:, -1] for subset in datasets])
                    
                    # Update best split if this is better
                    if info_gain > best_split["info_gain"]:
                        best_split["feature_index"] = feature_index
                        best_split["thresholds"] = thresholds
                        best_split["info_gain"] = info_gain
                        best_split["datasets"] = datasets
        
        return best_split

    def split(self, dataset, feature_index, thresholds):
        thresholds = sorted(thresholds)
        feature_values = dataset[:, feature_index]
        
        subsets = []
        # Handle data below first threshold
        mask = feature_values <= thresholds[0]
        subsets.append(dataset[mask])
        
        # Handle data between thresholds
        for i in range(1, len(thresholds)):
            mask = (feature_values > thresholds[i-1]) & (feature_values <= thresholds[i])
            subsets.append(dataset[mask])
        
        # Handle data above last threshold
        mask = feature_values > thresholds[-1]
        subsets.append(dataset[mask])
        
        return subsets

    def information_gain(self, parent, children, mode=None):
        # Use the class criterion if mode not specified
        if mode is None:
            mode = self.criterion
            
        # Weight of each child node
        weights = [len(child) / len(parent) for child in children]
        
        # Calculate entropy or gini for parent
        if mode == "gini":
            parent_impurity = self.gini_index(parent)
            children_impurity = [self.gini_index(child) for child in children]
        else:
            parent_impurity = self.entropy(parent)
            children_impurity = [self.entropy(child) for child in children]
        
        # Calculate gain
        gain = parent_impurity - sum(w * impurity for w, impurity in zip(weights, children_impurity))
        
        return gain

    def entropy(self, y):
        # Get frequency of each class
        _, counts = np.unique(y, return_counts=True)
        probabilities = counts / len(y)
        
        # Calculate entropy
        entropy = -np.sum([p * np.log2(p) for p in probabilities if p > 0])
        
        return entropy

    def gini_index(self, y):
        # Get frequency of each class
        _, counts = np.unique(y, return_counts=True)
        probabilities = counts / len(y)
        
        # Calculate Gini index
        gini = 1 - np.sum([p**2 for p in probabilities])
        
        return gini

    def calculate_leaf_value(self, Y):
        """Return the most common class in the leaf node"""
        return max(set(Y), key=list(Y).count)

    def print_tree(self, tree=None, indent=""):
        if tree is None:
            tree = self.root
        
        # If it's a leaf node
        if tree.value is not None:
            print(f"{indent}Predict: {tree.value}")
            return
        
        # Print decision node
        feature_name = self.feature_names[tree.feature_index] if self.feature_names else f"Feature {tree.feature_index}"
        print(f"{indent}{feature_name} with thresholds {tree.thresholds} (gain: {tree.info_gain:.4f})")
        
        # Print children
        for i, child in enumerate(tree.children):
            if i == 0:
                print(f"{indent}├── If {feature_name} <= {tree.thresholds[0]}:")
            elif i == len(tree.children) - 1:
                print(f"{indent}└── If {feature_name} > {tree.thresholds[-1]}:")
            else:
                threshold_lower = tree.thresholds[i-1]
                threshold_upper = tree.thresholds[i]
                print(f"{indent}├── If {threshold_lower} < {feature_name} <= {threshold_upper}:")
            
            self.print_tree(child, indent + "│   ")

    def fit(self, X, Y):
        # Store feature names if X is a DataFrame
        if isinstance(X, pd.DataFrame):
            self.feature_names = X.columns.tolist()
            X = X.values
        else:
            self.feature_names = [f"Feature_{i}" for i in range(X.shape[1])]
            
        if isinstance(Y, pd.Series):
            Y = Y.values
        
        # Create dataset by combining X and Y
        dataset = np.hstack((X, Y.reshape(-1, 1)))
        
        # Build the tree
        self.root = self.build_tree(dataset)
        
        return self

    def predict(self, X):
        # Convert to numpy array if needed
        if isinstance(X, pd.DataFrame):
            X = X.values
        
        # Make predictions for each sample
        return np.array([self.make_prediction(x, self.root) for x in X])

    def make_prediction(self, x, tree):
        if tree.value is not None:
            return tree.value

        feature_val = x[tree.feature_index]

        for i, threshold in enumerate(tree.thresholds):
            if feature_val <= threshold:
                return self.make_prediction(x, tree.children[i])

        return self.make_prediction(x, tree.children[-1])
    
    def visualize_tree(self, output_file="decision_tree_visualization"):
        """
        Create a visualization of the decision tree using graphviz.
        
        Args:
            output_file: Name of output file (without extension)
        """
        # Function to traverse the tree and build the dot graph
        def add_nodes_edges(tree, parent_id, dot, node_id=0):
            if tree.value is not None:  # Leaf node
                label = f"Predict: {tree.value}"
                dot.node(str(node_id), label, shape="box", style="filled", color="lightblue")
                if parent_id is not None:
                    dot.edge(str(parent_id), str(node_id))
                return node_id + 1
            
            # Decision node
            feature_name = self.feature_names[tree.feature_index] if self.feature_names else f"Feature {tree.feature_index}"
            label = f"{feature_name}\nGain: {tree.info_gain:.4f}"
            dot.node(str(node_id), label, shape="ellipse", style="filled", color="lightgreen")
            
            if parent_id is not None:
                dot.edge(str(parent_id), str(node_id))
            
            current_id = node_id
            next_id = node_id + 1
            
            # Add children nodes
            for i, child in enumerate(tree.children):
                if i == 0:
                    edge_label = f"≤ {tree.thresholds[0]}"
                elif i == len(tree.children) - 1:
                    edge_label = f"> {tree.thresholds[-1]}"
                else:
                    edge_label = f"{tree.thresholds[i-1]} < x ≤ {tree.thresholds[i]}"
                
                dot.edge(str(current_id), str(next_id), label=edge_label)
                next_id = add_nodes_edges(child, next_id, dot, next_id)
            
            return next_id
        
        # Create the dot graph
        dot = graphviz.Digraph(format='png')
        dot.attr('node', shape='box')
        
        # Add nodes and edges
        add_nodes_edges(self.root, None, dot)
        
        # Render the graph
        dot.render(output_file, cleanup=True)
        
        return dot

```

## Train and Visulization

```python
# Train both our custom tree and scikit-learn's tree for comparison
# Our implementation
custom_dt = DecisionTreeClassifier(min_sample_split=10, max_depth=3, num_splits=2, criterion="entropy")
custom_dt.fit(X_train, y_train)

# Make predictions
y_train_pred_custom = custom_dt.predict(X_train)
y_test_pred_custom = custom_dt.predict(X_test)

# Calculate accuracy
train_accuracy_custom = sum(y_train_pred_custom == y_train) / len(y_train)
test_accuracy_custom = sum(y_test_pred_custom == y_test) / len(y_test)

print(f"\nCustom Decision Tree - Training Accuracy: {train_accuracy_custom:.4f}")
print(f"Custom Decision Tree - Testing Accuracy: {test_accuracy_custom:.4f}")

# Visualize our custom tree
try:
    custom_dt.visualize_tree("custom_decision_tree")
    print("Custom tree visualization saved as 'custom_decision_tree.png'")
except Exception as e:
    print(f"Could not visualize custom tree: {e}")

# For comparison - sklearn implementation
from sklearn.tree import DecisionTreeClassifier as SklearnDT
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Train with similar parameters 
sklearn_dt = SklearnDT(min_samples_split=10, max_depth=3, criterion="entropy", random_state=42)
sklearn_dt.fit(X_train, y_train)

# Make predictions
y_train_pred_sklearn = sklearn_dt.predict(X_train)
y_test_pred_sklearn = sklearn_dt.predict(X_test)

# Calculate accuracy
train_accuracy_sklearn = accuracy_score(y_train, y_train_pred_sklearn)
test_accuracy_sklearn = accuracy_score(y_test, y_test_pred_sklearn)

print(f"\nScikit-learn Decision Tree - Training Accuracy: {train_accuracy_sklearn:.4f}")
print(f"Scikit-learn Decision Tree - Testing Accuracy: {test_accuracy_sklearn:.4f}")

# Create confusion matrix for test predictions
conf_matrix = confusion_matrix(y_test, y_test_pred_sklearn)

# More detailed evaluation metrics
print("\nClassification Report (Scikit-learn model):")
print(classification_report(y_test, y_test_pred_sklearn))

print("\nConfusion Matrix (Scikit-learn model):")
print(conf_matrix)

# Visualize the scikit-learn tree
plt.figure(figsize=(20, 10))
tree.plot_tree(sklearn_dt, feature_names=X.columns, class_names=['Dissatisfied', 'Satisfied'], filled=True, rounded=True)
plt.savefig('sklearn_decision_tree.png', dpi=300, bbox_inches='tight')
print("Scikit-learn tree visualization saved as 'sklearn_decision_tree.png'")

# Feature importance analysis
importances = sklearn_dt.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(12, 8))
plt.title('Feature Importance')
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), [X.columns[i] for i in indices], rotation=90)
plt.tight_layout()
plt.savefig('feature_importance_sklearn.png')
print("Feature importance visualization saved as 'feature_importance_sklearn.png'")

# Experiment with different hyperparameters to check for overfitting
depths = [2, 3, 5, 7, 10, 15]
min_samples = [5, 10, 20, 50, 100]

# Store results
results = []

# Test different hyperparameters
for depth in depths:
    for sample in min_samples:
        model = SklearnDT(max_depth=depth, min_samples_split=sample, random_state=42)
        model.fit(X_train, y_train)
        
        train_acc = accuracy_score(y_train, model.predict(X_train))
        test_acc = accuracy_score(y_test, model.predict(X_test))
        
        results.append({
            'max_depth': depth,
            'min_samples_split': sample,
            'train_accuracy': train_acc,
            'test_accuracy': test_acc,
            'difference': train_acc - test_acc  # Measure of overfitting
        })

# Convert to DataFrame for easier analysis
results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='test_accuracy', ascending=False)

print("\nTop 5 models by test accuracy:")
print(results_df.head())

# Visualize overfitting analysis
plt.figure(figsize=(12, 8))
for depth in depths:
    subset = results_df[results_df['max_depth'] == depth]
    plt.plot(subset['min_samples_split'], subset['train_accuracy'], 'o-', label=f'Train (depth={depth})')
    plt.plot(subset['min_samples_split'], subset['test_accuracy'], 's--', label=f'Test (depth={depth})')

plt.xlabel('Min Samples Split')
plt.ylabel('Accuracy')
plt.title('Model Performance vs Hyperparameters')
plt.legend()
plt.xscale('log')
plt.grid(True, alpha=0.3)
plt.savefig('hyperparameter_tuning.png')
print("Hyperparameter tuning analysis saved as 'hyperparameter_tuning.png'")
```
